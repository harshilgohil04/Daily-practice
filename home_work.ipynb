```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
sns.get_dataset_names()
```

```python
data=sns.load_dataset("mpg")
data
```

```python
data1=data.isnull().sum()
data1
```

```python
data.dropna(inplace= True)
data
```

```python
data2=pd.get_dummies(data,columns=["cylinders","model_year","origin"],drop_first=True)
data2.head()
```

```python
x=data2[["mpg","horsepower","weight",
        "acceleration","cylinders_4","cylinders_5","cylinders_6","cylinders_8","model_year_75","model_year_76","model_year_77",
        "model_year_78","model_year_79","model_year_80","model_year_81","model_year_82","origin_japan","origin_usa"]]


y=data2["displacement"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=150)
```

```python
x_train
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```
