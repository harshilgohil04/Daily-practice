is it easier to get a job for a female ?

```python
len(df.loc[(df.gender == 'Female') & (df.placement_status == 'Not Placed')]) / len(df.loc[(df.gender == 'Female')])
```

For Male ?


```python
len(df.loc[(df.gender == 'Male') & (df.placement_status == 'Not Placed')]) / len(df.loc[(df.gender == 'Male')])
```

do males earn higher than female ?

```python
male_salary=df.loc[(df.gender == 'Male')].salary.sum() / len(df.loc[(df.gender == 'Male')])
female_salary=df.loc[(df.gender == 'Female')].salary.sum() / len(df.loc[(df.gender == 'Female')])
print(male_salary)
print(female_salary)
```
